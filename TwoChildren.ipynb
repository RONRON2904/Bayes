Librairies python : pandas et numpy   
Numpy pour générer un tableau de deux entiers aléatoires (0 ou 1) de manière équiprobable   
Pandas pour rendre les données plus lisibles avec des dataframe et noms de colonnes.

```python
import numpy as np
import pandas as pd
```

On génère 10000 couples avec leurs deux enfants.   
On peut donc avoir soit (Boy, Boy), soit (Girl, Girl) soit (Boy, Girl) soit (Girl, Boy)

```python
children = pd.DataFrame(np.random.randint(0, 2, size=(10000, 2)), columns=['FirstChild', 'SecondChild'])
```

Les données sont rendues plus lisibles en transformant les 0 en garcon et les 1 en fille

```python
children['FirstChild'] = children['FirstChild'].apply(lambda x: 'Boy' if x == 0 else 'Girl')
children['SecondChild'] = children['SecondChild'].apply(lambda x: 'Boy' if x == 0 else 'Girl')
children
```

Seuls les couples ayant au moins un garcon nous intéresse.   
La condition logique est donc : le premier enfant est un garcon OU le second est un garcon.    
Ce n'est pas un OU exclusif, les deux enfants peuvent être des garcons.

```python
at_least_one_boy = children[(children['FirstChild'] == 'Boy') | (children['SecondChild'] == 'Boy')]
at_least_one_boy.reset_index()
```

Sur 10000 couples, 7484 ont au moins un garcon. On s'intéresse maintenant à la proportion de ceux qui ont deux enfant parmi ces 7484 couples.

```python
two_boys_freq = at_least_one_boy[at_least_one_boy['FirstChild'] == at_least_one_boy['SecondChild']].shape[0]/at_least_one_boy.shape[0]
print("Proportion des couples ayant 2 garcons sachant que l'un des deux enfants est un garcon", two_boys_freq)
```

Parmi les couples ayant deux enfants dont l'un d'eux est un garcon, 32.95% est la proportion de ceux qui ont deux enfants.

42% != 32.95%
